Examples of integrating a differential 1-form over 1-simplices

```python
import numpy as np
import matplotlib.pyplot as plt

# Set font to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'

# Define the vector field components with a = 1 and b = 1
def vector_field(x, y, a=1, b=1):
    u = -b * y  # dx component of the vector field
    v = a * x   # dy component of the vector field
    return u, v

# Define the vertices for the edges
vertices = [
    (np.array([1, 2]), np.array([-1, 2])), 
    (np.array([-1, -2]), np.array([-1, 1])), 
    (np.array([-2, 1]), np.array([-2, -2])), 
    (np.array([0, -2]), np.array([1, -2])), 
    (np.array([1, 0]), np.array([1, 1])) 
]

# Define colors for each edge
colors = ['red', 'green', 'blue', 'purple', 'orange']

# Set the parameters a and b
a = 3
b = 0

# Create a grid of points for plotting the vector field
x, y = np.meshgrid(np.linspace(-3, 3, 20), np.linspace(-3, 3, 20))
u, v = vector_field(x, y, a, b)  # Use the modified a and b

# Plot the vector field
plt.figure(figsize=(8, 8))
plt.quiver(x, y, u, v, color='blue', alpha=0.5)

# Plot each edge and calculate the simplicial 1-form value
for i, (v0, v1) in enumerate(vertices):
    color = colors[i]

    # Plot the edge
    plt.plot([v0[0], v1[0]], [v0[1], v1[1]], color=color, linewidth=2, label=fr'$\sigma^1_{{{i + 1}}}$')

    # Mark the vertices
    plt.plot(v0[0], v0[1], 'o', color=color)
    plt.plot(v1[0], v1[1], 'o', color=color)

    # Calculate the midpoint for the arrow
    midpoint = (v0 + v1) / 2
    arrow_direction = (v1 - v0) / 8  # Adjust the scale for visibility

    # Show an arrow indicating direction at the midpoint
    plt.arrow(midpoint[0], midpoint[1], arrow_direction[0], arrow_direction[1],
              head_width=0.1, head_length=0.3, fc=color, ec=color)

    # Calculate the value of the simplicial 1-form on the edge using the correct integral
    integral_value = a * (v1[0] * (v1[1] - v0[1])) - b * (v1[1] * (v1[0] - v0[0]))
    plt.text(midpoint[0], midpoint[1] - 0.3, f'{integral_value:.2f}', fontsize=18, color=color)

# Set plot limits and labels
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.xlabel(r'$x_1$', fontsize=18)
plt.ylabel(r'$x_2$', fontsize=18)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(False)
plt.legend(fontsize=18)
plt.title(r'Integrating differential 1-form $\omega^1$ over 1-simplices $\sigma^1_i$', fontsize=20)
plt.text(-2.7, 2.7, rf'$\omega^1(x_1,x_2)={{{a}}}x_1dx^2-{{{b}}}x_2dx^1$', fontsize=18, color='black', verticalalignment='top',bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.5'))

# Display the plot
plt.show()
```
